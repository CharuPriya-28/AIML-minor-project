```python
# ðŸ“š Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# âœ… Load the dataset (Capital H!)
df = pd.read_csv('/content/Housing.csv')
print("âœ… Data loaded successfully!")

# âœ… Define price ranges
bins = [0, 2500000, 5000000, 7500000, 10000000, df['price'].max()]
labels = ['0-25 lakhs', '26-50 lakhs', '51-75 lakhs', '76-100 lakhs', '>100 lakhs']

# âœ… Create new column for price ranges
df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels, right=True)

# âœ… Count houses in each range
price_counts = df['price_range'].value_counts().sort_index()
print("\nðŸ“Š Houses in each price range:\n")
print(price_counts)

# âœ… Visualize with a line chart
plt.figure(figsize=(10, 6))
plt.plot(labels, price_counts, marker='o', linestyle='-', color='teal')
plt.title('Number of Houses in Different Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Houses')
plt.grid(True)
plt.show()
```

```python
# ðŸ“š Import libraries
import pandas as pd

# âœ… Load dataset
df = pd.read_csv('/content/Housing.csv')
print("âœ… Data loaded successfully!")

# âœ… Check columns and unique prefarea values
print("\nðŸ“‚ Columns:", df.columns)
print("\nðŸ” Unique prefarea values:", df['prefarea'].unique())

# âœ… Convert prefarea to lower case for consistency (optional)
df['prefarea'] = df['prefarea'].str.lower()

# âœ… Filter: <5000 sqft & no prefarea
group1 = df[(df['area'] < 5000) & (df['prefarea'] == 'no')]

# âœ… Filter: >5000 sqft & prefarea
group2 = df[(df['area'] > 5000) & (df['prefarea'] == 'yes')]

# âœ… Calculate average prices
avg_price_group1 = group1['price'].mean()
avg_price_group2 = group2['price'].mean()

# âœ… Calculate price gap
price_gap = avg_price_group2 - avg_price_group1

print(f"\nðŸ  Avg price (<5000 sqft & no prefarea): {avg_price_group1:,.0f}")
print(f"ðŸ¡ Avg price (>5000 sqft & prefarea): {avg_price_group2:,.0f}")
print(f"\nðŸ’° Estimated price gap: {price_gap:,.0f}")
```
